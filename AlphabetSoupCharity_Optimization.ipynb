## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

# Import and read the charity_data.csv.
import pandas as pd
application_df = pd.read_csv("https://static.bc-edx.com/data/dl-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Make a copy of the DF before droping any columns
application_df_copy = application_df.copy()

# Show a snippet of the copy DF
application_df_copy.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
# application_df_copy = application_df_copy.drop(columns=['EIN', 'NAME', 'ORGANIZATION', 'STATUS', 'SPECIAL_CONSIDERATIONS'])
application_df_copy = application_df_copy.drop(columns=['EIN', 'NAME'])

# Show the df without the dropped columns
application_df_copy.head()
```

```python
# Determine the number of unique values in each column.
unique_values = application_df_copy.nunique()

# Display the unique values per column
print(unique_values)
```

```python
# Look at APPLICATION_TYPE value counts to identify and replace with "Other"
application_count = application_df_copy['APPLICATION_TYPE'].value_counts()

# Show the data
application_count
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
cutoff_value = 500
application_types_to_replace = application_count[application_count < cutoff_value].index.tolist()

# Replace in dataframe
for app in application_types_to_replace:
    application_df_copy['APPLICATION_TYPE'] = application_df_copy['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure replacement was successful
application_df_copy['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts to identify and replace with "Other"
classification_count = application_df_copy['CLASSIFICATION'].value_counts()

# Show the data
classification_count
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
classification_gt1 = classification_count[classification_count>1]

# Show the new count data
classification_gt1
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
cutoff_cv = 1000
classifications_to_replace = classification_count[classification_count < cutoff_cv].index.tolist()

# Replace in dataframe
for cls in classifications_to_replace:
    application_df_copy['CLASSIFICATION'] = application_df_copy['CLASSIFICATION'].replace(cls,"Other")

# Check to make sure replacement was successful
application_df_copy['CLASSIFICATION'].value_counts()
```

```python
# Look at INCOME_AMT value counts to identify and replace with "Other"
income_count = application_df_copy['INCOME_AMT'].value_counts()

# Show the data
income_count
```

```python
# Look at ASK_AMT value counts to identify and replace with "Other"
ask_count = application_df_copy['ASK_AMT'].value_counts()

# Show the data
ask_count
```

```python
# You may find it helpful to look at ASK_AMT value counts >1
ask_count_gt3 = ask_count[ask_count>3]

# Show the new count data
ask_count_gt3
```

```python
# Choose a cutoff value and create a list of ASK_AMT values to be replaced
# use the variable name `ask_to_replace`
cutoff_ask = 4
ask_to_replace = ask_count[ask_count < cutoff_ask].index.tolist()

# Replace in dataframe
for record in ask_to_replace:
    application_df_copy['ASK_AMT'] = application_df_copy['ASK_AMT'].replace(record,"Other")

# Check to make sure replacement was successful
application_df_copy['ASK_AMT'].value_counts()

```

```python
# Make a copy of the DF before turning categorical values into numeric
asoup_df = application_df_copy.copy()

# Show a snippet of the copy DF
asoup_df.head()
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
asoup_df = pd.get_dummies(asoup_df, dtype=int)

# Show the head of the DF
asoup_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
# Define features set
# Make a copy of the dataset
X = asoup_df.copy()

# Drop the column
X.drop(columns=['IS_SUCCESSFUL'], axis=1, inplace=True)

# Display the values the features columns
X.head()
```

```python
# Define Target Vector
y = asoup_df['IS_SUCCESSFUL'].values

# Show first 5 values in the array
y[:5]
```

```python
# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
# Transform the training data
X_train_scaled = X_scaler.transform(X_train)

# Transform the testing data
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, BatchNormalization, LeakyReLU, ELU
from tensorflow.keras.regularizers import l2
from tensorflow.keras.callbacks import EarlyStopping, LearningRateScheduler, ModelCheckpoint
from tensorflow.keras.optimizers import Adam, SGD, RMSprop, Nadam
from tensorflow.keras.activations import swish, relu
```

```python
# Define a learning rate scheduler
def scheduler(epoch, lr):
    if epoch < 10:
        return lr
    else:
        return lr * tf.math.exp(-0.1)
```

```python
# Define the model
nn = tf.keras.models.Sequential()

# Input layer with swish activation and L2 regularization
nn.add(Dense(units=100, kernel_regularizer=l2(0.001), input_dim=X_train_scaled.shape[1], activation='swish'))
nn.add(BatchNormalization())
nn.add(Dropout(0.2))

# Additional hidden layers with relu and L2 regularization
nn.add(Dense(units=80, activation='relu', kernel_regularizer=l2(0.001)))
nn.add(BatchNormalization())
nn.add(Dropout(0.2))

# Second Hidden Layer with relu and L2 regularization
nn.add(Dense(units=60, activation='relu', kernel_regularizer=l2(0.001)))
nn.add(Dropout(0.2))

# Third Hidden Layer with relu and L2 regularization
nn.add(Dense(units=60, activation='relu', kernel_regularizer=l2(0.001)))
nn.add(Dropout(0.2))

# Fourth Hidden Layer with relu and L2 regularization
nn.add(Dense(units=60, activation='relu', kernel_regularizer=l2(0.001)))
nn.add(Dropout(0.2))

# Fifth Hidden Layer with relu and L2 regularization
nn.add(Dense(units=60, activation='relu', kernel_regularizer=l2(0.001)))
nn.add(Dropout(0.2))

# Output layer
nn.add(Dense(units=1, activation='sigmoid'))

nn.summary()
```

```python
# Compile the model
nn.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Other optimizers used prior to just keeping Adam
# nn.compile(optimizer='SGD', loss='binary_crossentropy', metrics=['accuracy'])
# nn.compile(optimizer=RMSprop(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])
# nn.compile(optimizer=Nadam(learning_rate=0.001), loss='binary_crossentropy', metrics=['accuracy'])


```

```python
# Define callbacks
checkpoint_callback = ModelCheckpoint(filepath='checkpoint_weights.h5', save_weights_only=True, save_freq='epoch')
early_stopping = EarlyStopping(monitor='val_loss', patience=10)
lr_scheduler = LearningRateScheduler(scheduler)
```

```python
# Train the model and use the callbacks so that we can stop  the validation if the training stops improving
fit_model = nn.fit(X_train_scaled, y_train, epochs=100, validation_split=0.15, validation_data=(X_test_scaled, y_test), callbacks=[checkpoint_callback, early_stopping, lr_scheduler])

```
